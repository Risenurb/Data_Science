**Data Preparation**

```python
# renaming columns
df = df.rename(columns={"Unnamed: 2":"Month"})
```

```python
df[[#'IncidentNumber', 
    'DateOfCall', 'Month', 'CalYear', 
    'TimeOfCall',
       'HourOfCall', 'IncidentGroup', 
    #'StopCodeDescription',
     #  'SpecialServiceType', 
    'PropertyCategory', 
    #'PropertyType',
       'AddressQualifier', 'Postcode_full', 'Postcode_district', 'UPRN',
       'USRN', 'IncGeo_BoroughCode', 'IncGeo_BoroughName', 'ProperCase',
       'IncGeo_WardCode', 'IncGeo_WardName', 'IncGeo_WardNameNew', 'Easting_m',
       'Northing_m', 'Easting_rounded', 'Northing_rounded', 'Latitude',
       'Longitude', 'FRS', 'IncidentStationGround',
       'FirstPumpArriving_AttendanceTime',
       'FirstPumpArriving_DeployedFromStation',
       'SecondPumpArriving_AttendanceTime',
       'SecondPumpArriving_DeployedFromStation',
       'NumStationsWithPumpsAttending', 'NumPumpsAttending', 'PumpCount',
       'PumpHoursRoundUp', 'NotionalCost']]
```

```python
#missing values
```

```python
df.isna().sum()/df.shape[0]*100
```

```python
# duplicated values

df.loc[~df.duplicated()]
```

```python
# find not duplicates in the columns
df =df.loc[~df.duplicated(subset=["Month","CalYear","TimeOfCall","IncidentGroup"])]\
      .reset_index(drop=True).copy()
```

```python
df.query('PropertyCategory =="Non Residential"')
```

```python
df.shape
```

**Feature Understanding**

```python
ax =df["CalYear"].value_counts().plot(kind="bar", title="Dataset Years")
ax.set_xlabel("Year")
ax.set_ylabel("Quantity of records")
```

```python
# look at distribution . Ideally should be looked at each feature
df["FirstPumpArriving_AttendanceTime"].plot(kind="hist", bins=20)
```

```python
# kde plot
df["FirstPumpArriving_AttendanceTime"].plot(kind="kde")
```

**Feature relationship**

```python
df.head(4)
```

```python
ax=df.plot(kind="scatter",y="NotionalCost",x="PumpCount",title="Cost vs Pumps")
ax.set_xlabel("Pumps")
ax.set_ylabel("Cost")
plt.show
```

```python
# in the seaborn we can add hue

sns.scatterplot(y="NotionalCost",
                x="PumpCount",
                hue="IncidentGroup",
                data=df)
```
